## A) Interpolate the points: 

$p_1$ = (5.4, 3.2)
$p2_i$ = (9.5, 0.7)
$p_3$ = (12.3, -3.6)

```python
p1 = (5.4, 3.2)
p2_i = (9.5, 0.7)
p3 = (12.3, -3.6)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from ipywidgets import interact
```

```python

def update_plot(p2_x, p2_y):
    x_coords = np.array([p1[0], p2_x, p3[0]])
    y_coords = np.array([p1[1], p2_y, p3[1]])

    a, b, c = np.polyfit(x_coords, y_coords, 2)

    plt.figure(figsize=(10, 6))
    plt.scatter(x_coords, y_coords, color="red", label="Puntos")

    x = np.linspace(2, 15, 300)
    y = a*x**2 + b*x + c
    plt.plot(x, y, color="blue", label=f"Parábola: y = {a:.2f}x² + {b:.2f}x + {c:.2f}")

    plt.xlabel("X")
    plt.ylabel("Y")
    plt.title("Interpolación de Parábola en Tiempo Real")
    plt.grid()
    plt.legend()
    plt.show()
```

```python
_ = interact(update_plot, p2_x=(5.4, 12.3, 0.1), p2_y=(-10.0, 10.0, 0.1))
```
